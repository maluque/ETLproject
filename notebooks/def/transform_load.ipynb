### Clean weird characters

```python
for index, row in product_df.iterrows():
    description = row['description']
    if not description.isascii():
        print(f"Row {index}: Description contains special characters - {description}")
```

```python
product_df['description'] = product_df['description'].apply(lambda x: x.encode('utf-8', 'ignore').decode('utf-8'))


for index, row in product_df.iterrows():
    description = row['description']
    if not description.isascii():
        print(f"Row {index}: Description contains special characters - {description}")
        
        # Replace non-ASCII characters with a space (you can customize the replacement)
        cleaned_description = ''.join(char if char.isascii() else ' ' for char in description)
        
        # Update the DataFrame with the cleaned description
        product_df.at[index, 'description'] = cleaned_description

# Print the updated DataFrame
#print(product_df)
```
